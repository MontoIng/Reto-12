5. A través de un programa conectese a al menos 3 API's , obtenga el JSON, imprimalo y extraiga los pares de llave : valor.

```python
import requests
import json

def fetch_api_data(url, headers=None):
    """Realiza una solicitud GET a la API y devuelve los datos en formato JSON."""
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.json()
    else:
        return None

def extract_key_value_pairs(data, prefix=""): 
    """Extrae y muestra los pares clave-valor de un JSON anidado."""
    if isinstance(data, dict):
        for key, value in data.items():
            extract_key_value_pairs(value, prefix + key + ".")
    elif isinstance(data, list):
        for index, item in enumerate(data):
            extract_key_value_pairs(item, prefix + f"[{index}].")
    else:
        print(f"{prefix[:-1]}: {data}")

# URLs de las APIs
api_urls = {
    "Weather": "https://api.open-meteo.com/v1/forecast?latitude=40.71&longitude=-74.01&current_weather=true",
    "Bitcoin": "https://api.coindesk.com/v1/bpi/currentprice.json",
    "Random User": "https://randomuser.me/api/"
}

# Iterar sobre las APIs
for name, url in api_urls.items():
    print(f"\nConectando a la API de {name}...")
    data = fetch_api_data(url)
    
    if data:
        print(json.dumps(data, indent=4))  # Imprimir JSON completo
        print("\nPares clave-valor extraídos:")
        extract_key_value_pairs(data)  # Extraer clave-valor
    else:
        print("Error al obtener datos.")
```
